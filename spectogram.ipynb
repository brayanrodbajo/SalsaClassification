```python
import numpy as np
import librosa as lbr
import os
import librosa.display
import matplotlib.pyplot as plt
```

```python
WINDOW_SIZE = 4096
WINDOW_STRIDE = WINDOW_SIZE // 2
N_MELS = 128
MEL_KWARGS = {
    'n_fft': WINDOW_SIZE,
    'hop_length': WINDOW_STRIDE,
    'n_mels': N_MELS
}

filename = "D:/Documents/DL/FinalProject/SalsaClass/fragmentos-drive/fragmentos/Salsa/004-Vuelve - Andy Monta√±ez.mp3.wav"
filename = os.path.join(filename)
new_input, sample_rate = lbr.load(filename, mono=True)
features = lbr.feature.melspectrogram(new_input, **MEL_KWARGS)


# shape = (108,128)
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(lbr.power_to_db(features,
                                          ref=np.max),
                      y_axis='mel', fmax=8000,
                      x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram Salsa')
plt.tight_layout()
```
